```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
plt.figure(figsize=(12,5))

k1 = 1
k2 = 3
x = np.linspace(-10,10,300)

plt.plot(x, np.cos(k1*x), c='blue', label='y1')
plt.plot(x, np.cos(k2*x), c='green', label='y2')

plt.grid()
plt.legend()

plt.show()
```
