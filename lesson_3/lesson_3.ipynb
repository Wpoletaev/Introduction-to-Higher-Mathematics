#### 1.2) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
vector = input('Задайте вектор (введите координаты через пробел): ').strip()
if len(vector):
    vector = vector.split()
    result = 0
    for i in vector:
        try:
            i = int(i)
        except:
            print('Ошибка! Неправильный формат ввода!')
            result = None
            break
        result += i**2
    if result != None:
        result = np.sqrt(result)
        print(f'\nДлина ветора: {result}')
else:
    print('Ошибка! Вы ничего не ввели!')
```

#### 4.1) Решите систему уравнений:
y = x^2 – 1

exp(x) + x∙(1 – y) = 1


```python
# Графики:

rcParams['font.size'] = 8
plt.figure(figsize=(10,6))

x = np.linspace(-6,6,300)

y1 = x**2 - 1
y2 = 1 - (1 - np.exp(x))/x

plt.plot(x,y1,label='y1')
plt.plot(x,y2,label='y2')

plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.legend()

plt.xlim(-7,7)
plt.ylim(-3,25)

plt.show()
```

```python
# Решение системы уравнений:

from scipy.optimize import fsolve

def equations(params):
    x, y = params
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

X1, Y1 = fsolve(equations, (-2, 5))
X2, Y2 = fsolve(equations, (2, 5))
X3, Y3 = fsolve(equations, (5, 20))

print(f'x1: {X1}')
print(f'y1: {Y1}\n')

print(f'x2: {X2}')
print(f'y2: {Y2}\n')

print(f'x3: {X3}')
print(f'y3: {Y3}')
```

#### 4.2) Решите систему уравнений и неравенств:
y = x^2 – 1

exp(x) + x∙(1 – y) > 1


Из второго неравенства очевидно:

y < 1 - (1 - exp(x))/x

Откуда легко получить графическое решение:

```python
# Графическое решение:

rcParams['font.size'] = 8
plt.figure(figsize=(10,6))

x = np.linspace(-6,6,300)

y1 = x**2 - 1
y2 = 1 - (1 - np.exp(x))/x

plt.plot(x,y1,label='y1')
plt.plot(x,y2,label='y2')

plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.legend()

plt.xlim(-7,7)
plt.ylim(-3,25)

for i in range(len(x)):
    point_y = y1[i]
    point_x = x[i]
    if point_y < 1 - (1 - np.exp(point_x))/point_x:
        plt.scatter(point_x,point_y,c='red')

plt.show()
```
